Data from https://fred.stlouisfed.org/series/UNRATE/downloaddata

```python
import pandas as pd
url = "https://raw.githubusercontent.com/chemok78/unemployment-rates-visualisation/master/UNRATE.csv"
# pandas sets data type of DATE column to text (data type of object). Convert to datetime
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
# print the first 12 rows
print(unrate.head(12))
```

```python
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

```python
# assign first twelve rows in a variable(year 1948)
first_twelve = unrate[0:12]
plt.plot(first_twelve["DATE"], first_twelve["VALUE"])
plt.show()
```

```python
# rotate x-axis
plt.plot(first_twelve["DATE"],first_twelve["VALUE"])
plt.xticks(rotation=90)
plt.show()
```

```python
# axis labels and title
plt.plot(first_twelve["DATE"], first_twelve["VALUE"])
plt.xticks(rotation=90)
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

```python
# create a figure object
# figsize is in inches
fig = plt.figure(figsize=(12,12))
# add 2 rows and 1 column to figure object and place charts at position 1,2,3,4,5
# create Axes objects
ax1= fig.add_subplot(5,1,1)
ax2= fig.add_subplot(5,1,2)
ax3= fig.add_subplot(5,1,3)
ax4= fig.add_subplot(5,1,4)
ax5= fig.add_subplot(5,1,5)
# district rows for first and second year
first_year = unrate[0:12]
second_year = unrate[12:24]
third_year = unrate[24:36]
fourth_year = unrate[36:48]
fifth_year = unrate[48:60]
# create plots
ax1.plot(first_year["DATE"], first_year["VALUE"])
ax2.plot(second_year["DATE"], second_year["VALUE"])
ax3.plot(third_year["DATE"], third_year["VALUE"])
ax4.plot(fourth_year["DATE"], fourth_year["VALUE"])
ax5.plot(fifth_year["DATE"], fifth_year["VALUE"])
plt.show()
```

```python
# create new column with month only as integers
unrate['MONTH'] = unrate['DATE'].dt.month
fig = plt.figure(figsize=(10,6))
# extract years
first_year = unrate[0:12]
second_year = unrate[12:24]
third_year = unrate[24:36]
fourth_year = unrate[36:48]
fifth_year = unrate[48:60]
# plot first and second year in one chart by calling plt.plot repeatedly
plt.plot(first_year["MONTH"], first_year["VALUE"], c="red")
plt.plot(second_year["MONTH"], second_year["VALUE"], c="blue")
plt.plot(third_year["MONTH"], third_year["VALUE"], c="green")
plt.plot(fourth_year["MONTH"], fourth_year["VALUE"], c="orange")
plt.plot(fifth_year["MONTH"], fifth_year["VALUE"], c="black")
plt.show()
```
